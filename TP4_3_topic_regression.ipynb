```python
import csv
import numpy as np

with open('listado_de_cursos.csv', newline='') as f:
    reader = csv.reader(f)
    data = np.array(list(reader)).astype(int).T[0]

print(data[:10])
```

```python
import pandas as pd

df_revs = pd.DataFrame(data).set_index(0)
df_revs["title"]=""
df_revs["price"]=""
df_revs["price_val"]=0
df_revs["rating_avg"]=0
df_revs["rating_cnt"]=0
df_revs["rating_pos"]=0
df_revs["rating_neg"]=0
df_revs
```

```python
id = 2670212
udict = udemy.course_detail(id)
display(udict)
df_revs.loc[id,"title"] = udict["title"]
df_revs.loc[id,"price"] = udict["price"]
df_revs.loc[id,"price_val"] = float(udict["price_detail"]["amount"])
udict = udemy.course_reviews(id)
rating_list = np.array([float(x["rating"]) for x in udict['results']])
df_revs.loc[id,"rating_avg"] = np.mean(rating_list)
df_revs.loc[id,"rating_cnt"] = len(rating_list)
df_revs.loc[id,"rating_pos"] = sum(rating_list>=4)
df_revs.loc[id,"rating_neg"] = sum(rating_list<4)
df_revs.loc[id,"name"] = sum(rating_list<4)
df_revs.loc[id]
```

```python
from tqdm.notebook import tqdm

ids = df_revs.index

for id in tqdm(ids, total=len(ids)):
    udict = udemy.course_detail(id)
    try:
        df_revs.loc[id,"title"] = udict["title"]
        df_revs.loc[id,"price"] = udict["price"]
        if udict["is_paid"]==True:
            df_revs.loc[id,"price_val"] = float(udict["price_detail"]["amount"])
    except:
        print(id, end=" - ")
        display(udict)
    udict = udemy.course_reviews(id)
    rating_list = np.array([float(x["rating"]) for x in udict['results']])
    if udict["next"] is not None:
        print(f"el id {id} tiene más reviews para descargar")
        udict = udemy.course_reviews(id,page=2)
        np.append(rating_list,np.array([float(x["rating"]) for x in udict['results']]))
    df_revs.loc[id,"rating_avg"] = np.mean(rating_list)
    df_revs.loc[id,"rating_cnt"] = len(rating_list)
    df_revs.loc[id,"rating_pos"] = sum(rating_list>=4)
    df_revs.loc[id,"rating_neg"] = sum(rating_list<4)
    #df_revs.loc[id]
```
